# **4. 모델 학습**
+ YOLO 모델은 v5부터 n(nano), s(small), m(medium), l(large), xl(xlarge) 경량 및 소형 버전이 있음
  + 각 버전에 맞는 학습이 맞는 파라미터가 있으니 맞게 조정해야 함  
  + 안 그러면 학습을 할려고 할 때 제대로 학습을 진행이 안 될 수 있음
  + 코랩은 보통 무료버전이기 때문에 수집한 데이터셋과 모델 크기에 따라 학습을 할 수 있는 버전이 제한되어 있음(제공해주는 GPU에 따라 제한이 있음)
  + s버전까지 학습이 무사히 되지만 m 버전부터는 코랩에서 제공해주는 환경 제한으로 학습이 안 될 수 있음(주의!)   

```python
from ultralytics import YOLO

# 모델 불러오기
#model = YOLO("/content/ultralytics/ultralytics/cfg/models/12/yolo12s.yaml")

# 모델 학습 중단시 중단된 부분부터 학습 이어가기(4에포크 학습 중에 last.pt 파일 생성됨)
model = YOLO("/content/drive/MyDrive/yolov12s_train_result/weights/last.pt")

# 학습 파라미터
results = model.train(
    data= "/content/SP_dataset/data.yaml",
    epochs=100,
    batch=16,
    imgsz=640,
    scale=0.9,  # S:0.9; M:0.9; L:0.9; X:0.9
    mosaic=1.0,
    mixup=0.05,  # S:0.05; M:0.15; L:0.15; X:0.2
    copy_paste=0.15,  # S:0.15; M:0.4; L:0.5; X:0.6
    device=0,
    # 학습이 중단 되면 써야 되는 코드
    resume=True,

    # 학습 결과 저장
    project="/content/drive/MyDrive/",
    name="yolov12s_train_result",
    exist_ok=True
)
```

# **5. 모델 검증**
