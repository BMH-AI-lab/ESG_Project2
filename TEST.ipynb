```python
from ultralytics import YOLO
import cv2

# 1. 사전 학습된 YOLOv12 모델 로드 (n, s, m, l, x 크기 선택 가능)
model = YOLO(r"C:\ESG_Project2\yolov12s_train_result\weights\best.pt") 
# 2. 이미지/비디오에 대한 추론 수행
results = model(r"C:\Users\배명훈\Downloads\더러운 패널7.jpg") # 탐지할 이미지 경로 입력

# 3. 결과 시각화 및 저장
for result in results:
    # 결과가 그려진 이미지를 화면에 표시
    res_plotted = result.plot()
    cv2.imshow("YOLOv12 Detection", res_plotted)
    
    # 결과를 파일로 저장 (runs/detect/predict 폴더에 저장됨)
    result.save(filename='result.jpg')

    # 탐지된 클래스 ID, 이름, 신뢰도 출력
    for box in result.boxes:
        print(f"Class: {int(box.cls[0])} ({model.names[int(box.cls[0])]}), "
              f"Confidence: {float(box.conf[0]):.2f}, "
              f"Box: {box.xyxy[0]}")

cv2.waitKey(0)
cv2.destroyAllWindows()
```
